```python
import imageio as img
import numpy as np
import matplotlib.pyplot as plt

# Membaca gambar berwarna (RGB)
path = "/content/KKn_Pemandangan.jpeg"
imgRGB = img.imread(path)

# Konversi gambar ke grayscale menggunakan rumus standar
imgGrayscale = np.dot(imgRGB[...,:3], [0.2989, 0.5870, 0.1140])

# Hitung histogram dari gambar grayscale
histogram, bins = np.histogram(imgGrayscale.flatten(), bins=256, range=[0,256])

# Plot gambar grayscale dan histogram
plt.figure(figsize=(12,6))

# Gambar grayscale
plt.subplot(1,2,1)
plt.imshow(imgGrayscale, cmap='gray')
plt.title('Gambar Grayscale')
plt.axis('off')

# Histogram grayscale
plt.subplot(1,2,2)
plt.plot(histogram)
plt.title('Histogram Grayscale')
plt.xlabel('Intensitas Piksel (0-255)')
plt.ylabel('Jumlah Piksel')

plt.show()

# Total piksel untuk setiap intensitas
for i in range(256):
    print(f"Intensitas {i}: {histogram[i]} piksel")
```
